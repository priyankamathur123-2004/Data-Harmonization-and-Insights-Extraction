#  Perform basic statistical analysis to evaluate central tendencies (mean, median, mode).


```python
data = {'instant':[1,2,3,4,5],
       'dteday':[10,11,10,12,10],
       'session':[1,1,2,1,2],
       'year':[0,1,2,3,0,],
       'month':[12,3,4,5,5],
       'hour':[12,8,9,10,10,],
       'weekday':[1,4,5,6,7],
       'holiday':['false','false','false','false','false'],
       'temp':[0.24,0.32,0.22,0.23,0.32],
       'weathersit':[1,2,3,4,5,],
       }
df = pd.DataFrame(data)
```

```python
df
```

```python
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
mode_values = df.mode(numeric_only=True).iloc[0]  
# Categorical columns (session, weekday, weathersit)
categorical_columns = ['session', 'weekday', 'weathersit']
categorical_mode = {col: df[col].mode()[0] for col in categorical_columns}

# Display the results
print("Mean Values (Numerical Columns):")
print(mean_values)
print("\nMedian Values (Numerical Columns):")
print(median_values)
print("\nMode Values (Numerical Columns):")
print(mode_values)
print("\nCategorical Mode Values:")
print(categorical_mode)
```

# HANDLE OUTLIERS

```python
# Handle outliers for numerical columns
for column in ['temp', 'hum', 'windspeed']:
    q1 = merged_1_and_2[column].quantile(0.35)
    q3 = merged_1_and_2[column].quantile(0.55)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    # CAP OUTLIERS
    merged_1_and_2[column] = merged_1_and_2[column].clip(lower=lower_bound, upper=upper_bound)

print("\nOutliers handled successfully.")
    
```
